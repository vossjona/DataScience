### Store the dataset in a Table

##### In many cases the dataset to be analyzed is available as a .CSV (comma separated values) file, perhaps on the internet. To analyze the data 
##### using SQL, it first needs to be stored in the database.

##### We will first read the csv files  from the given url  into pandas dataframes

##### Next we will be using the  df.to_sql() function to convert each csv file  to a table in sqlite  with the csv data loaded in it.


```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCensusData.csv")
df.to_sql("CENSUS_DATA", con, if_exists='replace', index=False,method="multi")

df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCrimeData.csv")
df.to_sql("CHICAGO_CRIME_DATA", con, if_exists='replace', index=False, method="multi")

df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv")
df.to_sql("CHICAGO_PUBLIC_SCHOOLS_DATA", con, if_exists='replace', index=False, method="multi")
```

### Problem 4

##### What are the top 10 schools with the highest "Average Student Attendance"?


```python
%%sql
select SCHOOL_ID, NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE from CHICAGO_PUBLIC_SCHOOLS_DATA
order by AVERAGE_STUDENT_ATTENDANCE desc
limit 10;
```

### Problem 5

##### Retrieve the list of 5 Schools with the lowest Average Student Attendance sorted in ascending order based on attendance


```python
%%sql
select SCHOOL_ID, NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE from CHICAGO_PUBLIC_SCHOOLS_DATA
order by AVERAGE_STUDENT_ATTENDANCE asc
limit 5;
```

### Problem 6

##### Now remove the '%' sign from the above result set for Average Student Attendance column


```python
%sql SELECT Name_of_School, REPLACE(Average_Student_Attendance, '%', '') as Average_Student_Attendance_In_Percent \
     from CHICAGO_PUBLIC_SCHOOLS_DATA \
     order by Average_Student_Attendance \
     LIMIT 5
```

### Problem 7

##### Which Schools have Average Student Attendance lower than 70%?


```python
%%sql
SELECT Name_of_School, REPLACE(Average_Student_Attendance, '%', '') AS Average_Student_Attendance_In_Percent
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
WHERE CAST(Average_Student_Attendance_In_Percent as DOUBLE) < 70
ORDER BY Average_Student_Attendance_In_Percent
```

### Problem 8

##### Get the total College Enrollment for each Community Area


```python
%%sql
SELECT COMMUNITY_AREA_NUMBER, Community_Area_Name, SUM(COLLEGE_ENROLLMENT) AS TOTAL_ENROLLMENT, ROUND(AVG(College_Enrollment_Rate__), 2) AS AVG_ENROLLMENT_RATE
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
GROUP BY Community_Area_Name
ORDER BY TOTAL_ENROLLMENT DESC;
```

### Problem 9

##### Get the 5 Community Areas with the least total College Enrollment  sorted in ascending order


```python
%%sql
SELECT COMMUNITY_AREA_NUMBER, Community_Area_Name, SUM(COLLEGE_ENROLLMENT) AS TOTAL_ENROLLMENT
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
GROUP BY Community_Area_Name
ORDER BY TOTAL_ENROLLMENT
LIMIT 5;
```

### Problem 10

##### List 5 schools with lowest safety score.


```python
%%sql
SELECT SCHOOL_ID, NAME_OF_SCHOOL, SAFETY_SCORE
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
WHERE SAFETY_SCORE !='None'
ORDER BY SAFETY_SCORE
LIMIT 5;
```

### Problem 11

##### Get the hardship index for the community area which has Total College Enrollment of 7541


```python
# need to find hardship index first
%sql SELECT name,type FROM PRAGMA_TABLE_INFO('CENSUS_DATA') ORDER BY name ASC;
```

```python
%%sql
SELECT ES.COMMUNITY_AREA_NUMBER, ES.Community_Area_Name, ES.Total_Enrollment, CD.HARDSHIP_INDEX
FROM (
    SELECT CPSD.COMMUNITY_AREA_NUMBER, CPSD.Community_Area_Name, SUM(CPSD.COLLEGE_ENROLLMENT) AS Total_Enrollment
    FROM CHICAGO_PUBLIC_SCHOOLS_DATA CPSD
    GROUP BY COMMUNITY_AREA_NUMBER
) AS ES,CENSUS_DATA CD
WHERE CD.COMMUNITY_AREA_NUMBER = ES.COMMUNITY_AREA_NUMBER
AND ES.Total_Enrollment = 7541;

```

### Problem 12

##### Get the hardship index for the community area which has the highest value for Total College Enrollment


```python
%%sql
SELECT ES.COMMUNITY_AREA_NUMBER, ES.Community_Area_Name, ES.Total_Enrollment, CD.HARDSHIP_INDEX
FROM (
    SELECT CPSD.COMMUNITY_AREA_NUMBER, CPSD.Community_Area_Name, SUM(CPSD.COLLEGE_ENROLLMENT) AS Total_Enrollment
    FROM CHICAGO_PUBLIC_SCHOOLS_DATA CPSD
    GROUP BY COMMUNITY_AREA_NUMBER
) AS ES, CENSUS_DATA CD
WHERE CD.COMMUNITY_AREA_NUMBER = ES.COMMUNITY_AREA_NUMBER
AND ES.Total_Enrollment = (
    SELECT MAX(Total_Enrollment)
    FROM (
        SELECT COMMUNITY_AREA_NUMBER, SUM(COLLEGE_ENROLLMENT) AS Total_Enrollment
        FROM CHICAGO_PUBLIC_SCHOOLS_DATA
        GROUP BY COMMUNITY_AREA_NUMBER
    )
);
```

```python
# just to check:
%sql SELECT CPSD.COMMUNITY_AREA_NUMBER, CPSD.Community_Area_Name, SUM(CPSD.COLLEGE_ENROLLMENT) AS Total_Enrollment \
    FROM CHICAGO_PUBLIC_SCHOOLS_DATA CPSD \
    GROUP BY COMMUNITY_AREA_NUMBER \
    ORDER BY Total_Enrollment DESC;
```
